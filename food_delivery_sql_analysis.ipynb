# Uber Eats Merchant Operations Analysis using SQL

## Objective

```python
# Import libraries for data analysis and SQL database work
import pandas as pd
import sqlite3
```

## Load Dataset
The dataset contains Uber Eats merchant-level information including city, state, delivery time, review metrics, promotions, and pricing tiers.

```python
# Load the Uber Eats merchant dataset
df = pd.read_csv("data/Ubereat_US_Merchant.csv")

# Preview the first 5 rows
df.head()
```

## Inspect Dataset
Review the structure, columns, data types, and missing values before analysis.

```python
# Check dataset shape
print("Dataset shape:", df.shape)

# Check column names
print("\nColumns:")
print(df.columns.tolist())

# Check data types and non-null counts
print("\nDataset info:")
df.info()
```

```python
# Check missing values
print("Missing values by column:")
print(df.isnull().sum())
```

## Create SQLite Database
Load the dataset into a SQLite database so SQL queries can be used for analysis.

```python
# Create a SQLite database connection
conn = sqlite3.connect("ubereats_merchants.db")

# Write the dataframe into a SQL table
df.to_sql("merchants", conn, if_exists="replace", index=False)

print("Table created successfully.")
```

## Clean Delivery Time
The delivery time column is stored as text ranges such as "15–30 min".  
This is converted into an estimated numeric midpoint for analysis.

```python
# Create a copy of the dataset for cleaning
clean_df = df.copy()

# Function to convert text delivery time ranges into approximate midpoint minutes
def extract_delivery_minutes(value):
    if pd.isna(value):
        return None
    
    text = str(value).replace("–", "-").replace(" min", "").strip()
    
    if "-" in text:
        parts = text.split("-")
        try:
            low = float(parts[0].strip())
            high = float(parts[1].strip())
            return (low + high) / 2
        except:
            return None
    else:
        try:
            return float(text)
        except:
            return None

# Apply the function
clean_df["delivery_minutes"] = clean_df["delivery_time"].apply(extract_delivery_minutes)

# Fill missing promotions with 'No Promotion'
clean_df["promotion"] = clean_df["promotion"].fillna("No Promotion")

# Preview cleaned columns
clean_df[["delivery_time", "delivery_minutes", "promotion"]].head()
```

```python
# Reload cleaned data into SQL
clean_df.to_sql("merchants", conn, if_exists="replace", index=False)

print("Cleaned table written to SQL successfully.")
```

## Merchant Coverage Analysis

This section analyzes how Uber Eats merchants are distributed across cities and states.  
Understanding merchant coverage helps identify which regions have stronger platform presence.

```python
query = """
SELECT COUNT(*) AS total_merchants
FROM merchants;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT state,
       COUNT(*) AS merchant_count
FROM merchants
GROUP BY state
ORDER BY merchant_count DESC;
"""

state_counts = pd.read_sql_query(query, conn)
state_counts.head(10)
```

```python
query = """
SELECT state,
       COUNT(*) AS merchant_count
FROM merchants
GROUP BY state
ORDER BY merchant_count DESC;
"""

state_counts = pd.read_sql_query(query, conn)
state_counts.head(10)
```

### Interpretation

Merchant coverage varies significantly across states and cities.  
Locations with higher merchant counts may indicate stronger platform adoption and greater customer demand.

## Delivery Time Analysis

Delivery time is an important operational metric for food delivery platforms.  
This section analyzes average estimated delivery time across different locations.

```python
query = """
SELECT AVG(delivery_minutes) AS avg_delivery_minutes
FROM merchants
WHERE delivery_minutes IS NOT NULL;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT state,
       ROUND(AVG(delivery_minutes),2) AS avg_delivery_minutes
FROM merchants
WHERE delivery_minutes IS NOT NULL
GROUP BY state
ORDER BY avg_delivery_minutes ASC;
"""

delivery_by_state = pd.read_sql_query(query, conn)
delivery_by_state.head(10)
```

```python
query = """
SELECT state,
       ROUND(AVG(delivery_minutes),2) AS avg_delivery_minutes
FROM merchants
WHERE delivery_minutes IS NOT NULL
GROUP BY state
ORDER BY avg_delivery_minutes DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

### Interpretation

Average delivery time varies across states, which may reflect differences in merchant density, traffic conditions, or delivery logistics efficiency.

## Ratings and Reviews Analysis

Customer ratings and review counts provide insight into merchant performance and customer satisfaction.

```python
query = """
SELECT loc_name,
       city,
       state,
       review_rating,
       review_count
FROM merchants
WHERE review_rating IS NOT NULL
ORDER BY review_rating DESC, review_count DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT state,
       ROUND(AVG(review_rating),2) AS avg_rating
FROM merchants
WHERE review_rating IS NOT NULL
GROUP BY state
ORDER BY avg_rating DESC;
"""

rating_by_state = pd.read_sql_query(query, conn)
rating_by_state.head(10)
```

```python
query = """
SELECT loc_name,
       city,
       state,
       review_count,
       review_rating
FROM merchants
WHERE review_count IS NOT NULL
ORDER BY review_count DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

### Interpretation

Highly rated merchants with large review volumes likely represent strong performers on the platform.  
Average ratings by state provide insight into regional customer satisfaction patterns.

## Promotion Analysis

Promotions can help merchants attract customers and increase order volume.  
This section analyzes how frequently promotions are used across merchants.

```python
query = """
SELECT promotion,
       COUNT(*) AS merchant_count
FROM merchants
GROUP BY promotion
ORDER BY merchant_count DESC;
"""

promo_counts = pd.read_sql_query(query, conn)
promo_counts
```

```python
query = """
SELECT 
    CASE 
        WHEN promotion = 'No Promotion' THEN 'No Promotion'
        ELSE 'Promotion Used'
    END AS promotion_status,
    COUNT(*) AS merchant_count
FROM merchants
GROUP BY promotion_status;
"""

pd.read_sql_query(query, conn)
```

### Interpretation

Promotion usage varies across merchants.  
Promotions may be used as a strategy to attract customers, especially in competitive markets.

## Merchant Availability Analysis

Merchant availability impacts customer access to food delivery services.  
This section examines the number of merchants currently open versus closed.

```python
query = """
SELECT is_open,
       COUNT(*) AS merchant_count
FROM merchants
GROUP BY is_open;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT state,
       COUNT(*) AS open_merchants
FROM merchants
WHERE is_open = 1
GROUP BY state
ORDER BY open_merchants DESC;
"""

pd.read_sql_query(query, conn)
```

### Interpretation

Merchant availability provides insight into platform accessibility.  
Regions with more open merchants may offer customers greater choice and shorter delivery times.

## Price Bucket Analysis

Price buckets categorize merchants into different pricing tiers.  
This analysis examines how merchants are distributed across price levels.

```python
query = """
SELECT price_bucket,
       COUNT(*) AS merchant_count
FROM merchants
GROUP BY price_bucket
ORDER BY merchant_count DESC;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT price_bucket,
       ROUND(AVG(review_rating),2) AS avg_rating
FROM merchants
WHERE review_rating IS NOT NULL
GROUP BY price_bucket
ORDER BY avg_rating DESC;
"""

pd.read_sql_query(query, conn)
```

### Interpretation

Price bucket analysis helps understand how merchants are distributed across different pricing tiers and whether higher-priced merchants tend to have higher customer ratings.
